```python
# Re-plotting the graph and attempting to display it directly.
import matplotlib.pyplot as plt

# Data from the image
quarters = ['1분기', '2분기', '3분기', '4분기']
delinquency_rates = [7.6, 8.4, 8.9, 9.2]

# Plotting the data
plt.figure(figsize=(8, 5))
plt.bar(quarters, delinquency_rates, color='blue')
plt.xlabel('분기')
plt.ylabel('연체율 (%)')
plt.title('분기별 연체율 증가')

plt.ylim(7, 10)  # Setting y-axis limit for better visualization
for i, rate in enumerate(delinquency_rates):
    plt.text(i, rate + 0.2, f'{rate}%', ha='center')
plt.savefig('분기별 연체율 증가.png')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# 그래프 크기 설정
plt.figure(figsize=(15, 10))

# 각 컬럼에 대한 상자 그림 생성
for i, column in enumerate(df.columns, 1):
    plt.subplot(4, 4, i)  # 4x4 그리드에 그래프 배치
    sns.boxplot(data=df[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
# VALUE
# Boxplot을 보니 이상치들이 많아 보여서(현재가치가 높은 사람들이 많아보여서)
# 결측값을 평균값으로 대체하기보단 중위수로 대체하였다
```

```python
# VALUE 컬럼의 중앙값 계산 
value_median = df['VALUE'].median() # VALUE 컬럼의 결측값을 중앙값으로 대체 
df['VALUE'].fillna(value_median, inplace=True)
```

```python
# JOB별 CLAGE 평균값 데이터
jobs = ['Mgr', 'Office', 'Other', 'ProfExe', 'Sales', 'Self']
values = [111657.158228, 96266.731826, 88931.236313, 132361.913514, 115425.281250, 160603.296875]
# 막대그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(jobs, values, color='skyblue')
plt.xlabel('Job')
plt.ylabel('Average CLAGE')
plt.title('Average CLAGE by JOB')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('JOB 별 평균 VALUE 막대 그래프.png')
plt.show()
```

```python
# BAD: 대출금 상환 여부, 0은 상환, 1은 미상환
# LOAN: 대출금액($)
# MORTDUE: 저당금액($)
# VALUE: 현 재의 자산가치($)
# REASON: 대출사유, Homelmp(주택개량), 빚정리(debtcon), 기타(Other)
# JOB: 직업의종류(7가지)
# YOJ: 현재 직장에서의 근무년수(Year)
# DEROG: 불량사유보고횟수
# DELINQ: 신용한도
# CLAGE: 계좌유지 기간
# NINQ: 최근의 신용상태 조사수
# CLNO: 금융거래의 수
# DEBTINC: 수입에 대한 부채비율(%)

# YOJ의 205개의 결측치에 대해서는 YOJ가 현재 직장에서의 근무년수(Year) 이므로 직업별 YOJ의 평균으로 대체
df['YOJ'].value_counts()
```

```python
#JOB별 YOJ의 평균 계산 
job_yoj_mean = df.groupby('JOB')['YOJ'].transform('mean')
# YOJ 컬럼의 결측값을 JOB별 YOJ의 평균으로 대체 
df['YOJ'].fillna(job_yoj_mean, inplace=True)
```
